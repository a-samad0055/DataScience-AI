```python
#Solve following question by using conditional statements (10) 
# How many studies have an enrollment 
#greater than a certain threshold? 
# How many clinical trials have 'No Results Available'? 
# How many clinical trials are in an "Completed" and "Recruiting" status? 
# How many clinical trials are related to 'COVID-19'? 
# How many clinical trials started after January 1, 2020?
```

```python
# How many studies have an enrollment 

enrollment_count = df['Enrollment'].count()

print("Number of studies with enrollment data:", enrollment_count)
enrollment_count
```

```python
#greater than a certain threshold? 

threshold = 100 


enrollment_gt_threshold = (df['Enrollment'] > threshold).sum()
print(f"Number of studies with enrollment greater than {threshold}: {enrollment_gt_threshold}")

print(enrollment_gt_threshold)
```

```python
# How many clinical trials have 'No Results Available'? 

no_results_count = (df['Study Results'] == 'No Results Available').sum()
print("Number of clinical trials with 'No Results Available':", no_results_count)

no_results_count
```

```python
# How many clinical trials are in an "Completed" and "Recruiting" status? 


completed_and_recruiting_count = ((df['Status'] == 'Completed') & (df['Status'] == 'Recruiting')).sum()
print("Number of clinical trials in 'Completed' and 'Recruiting' status:", completed_and_recruiting_count)

completed_and_recruiting_count
```

```python
# How many clinical trials are related to 'COVID-19'? 


covid_trials_count = (df['Conditions'].str.contains('COVID-19', case=False, na=False)).sum()
print("Number of clinical trials related to 'COVID-19':", covid_trials_count)

covid_trials_count
```

```python
# How many clinical trials started after January 1, 2020?

trials_after_2020 = df[df['Start Date'] > '2020-01-01'].shape[0]
trials_after_2020
```

```python
from datetime import datetime


df['Start Date'] = pd.to_datetime(df['Start Date'])


start_date_threshold = datetime(2020, 1, 1)
trials_after_2020_count = (df['Start Date'] > start_date_threshold).sum()
print("Number of clinical trials started after January 1, 2020:", trials_after_2020_count)
```
